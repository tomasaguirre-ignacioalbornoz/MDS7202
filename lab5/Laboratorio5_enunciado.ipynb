````python
# Reshape DataFrame usando melt
df_melted = df_grades.melt(value_vars=['math score', 'reading score', 'writing score'],
                            var_name='subject', value_name='score')
df_melted['subject'] = df_melted['subject'].str.replace(' score', '')

df_melted
````

**Gr치fico de Caja:**

````python
fig_boxplot = px.box(
    df_melted,
    y='score',
    x='subject',
    title='Box plot of subject scores',
    color='subject'
)

fig_boxplot.update_traces(showlegend=True)
fig_boxplot.show()

````

**Distplot:**

````python
# use displot figurefactory

# Create a list of color codes for each subject
colors = {'math': 'blue', 'reading': 'orange', 'writing': 'green'}
df_melted['color'] = df_melted['subject'].map(colors)

# Create the displot using Plotly Figure Factory
fig_displot = ff.create_distplot(
    [df_melted[df_melted['subject'] == subject]['score'] for subject in df_melted['subject'].unique()],
    df_melted['subject'].unique(),
    colors=list(colors.values()),
    show_hist=False,  # Do not display histograms
    histnorm='probability',  # Set histogram normalization to 'probability'
    show_rug=False,  # Do not display rug plots
)

# Customize the legend title and labels
fig_displot.update_layout(
    title='Distribution of subject scores',
    xaxis_title='score',
    yaxis_title='density',
    legend_title_text='subject',
)

fig_displot.show()
````

**Histograma con Boxplots:**

````python
# Define the score bins to be like spected result
score_bins = [0, 1, 2, 3, 4, 6, 7]  


# Create histograms using Plotly Express
fig_hist = px.histogram(df_melted, x='score', color='subject',
                        marginal='box',  # Add boxplots as marginal plots                      
                        nbins=len(score_bins)*2,  # Number of bins is one less than the boundaries
                        #range_x=[min(score_bins), max(score_bins)], # Set the x-axis range
                        title='Histogram of subject scores with box plots')

# Customize the legend title and labels
fig_hist.update_layout(
    xaxis_title='score', # grade or score, here we call this variable score to maintain consistency
    yaxis_title='count',
    legend_title_text='subject',
    barmode ='group' # In 'group' mode, bars are placed beside each other
)


fig_hist.show()
````

**Histograma con Faceta:**

````python
# Create histograms with facets using Plotly Express
fig_facet_hist = px.histogram(
    df_melted,
    x='score',
    facet_row='subject',  # Create facets for each subject
    color='subject',  # Color by the 'subject' column
    nbins=len(score_bins)*2,
    title='Histogram with facets of subject scores',
    labels={'score': 'Score'},
)

# Customize the legend title and labels
fig_facet_hist.update_layout(
    xaxis_title='score',
    yaxis_title='count',
    legend_title_text='subject'
)

# Manually adjust the height of the figure to create more space between facets
# To avoid overlap between each of the subjects = ...
fig_facet_hist.update_layout(
    height=600  # Adjust the height as needed
)

fig_facet_hist.show()

````

### 1.3.4 Informaci칩n por Categor칤as [1 punto]

````python
# C치lculo de GPA
df_grades['GPA'] = df_grades[['math score', 'reading score', 'writing score']].mean(axis=1)
````

````python
# Mapeo parental level of education
def map_edu(education):
    if education == "some high school":
        return "school"
    elif education == "some college":
        return "school"
    elif education == "high school":
        return "school"
    elif education == "bachelor's degree":
        return "college"
    elif education == "associate's degree":
        return "college"
    elif education == "master's degree":
        return "postgraduate"
    
df_grades["simple parental level of education"] = df_grades["parental level of education"].map(map_edu) 
````

````python
# Agrupa por las columnas 'race/ethnicity' y ' simple parental level of education', y calcula el promedio de las notas
df_agregado_1 = df_grades.groupby(['race/ethnicity', 'simple parental level of education'])[['math score', 'reading score', 'writing score', "GPA"]].mean().reset_index()
````

````python
# Agrupa por las mismas columnas y cuenta los alumnos en cada grupo
df_agregado_2 = df_grades.groupby(['race/ethnicity', 'simple parental level of education']).size().reset_index(name='count')

# Concatena los DataFrames a lo largo de las filas
df_agregado = pd.concat([df_agregado_1, df_agregado_2['count']], axis=1)
````

````python
# Calcula el total de alumnos en todas las filas
total_students = df_agregado['count'].sum()

# Calcula el porcentaje y crea una nueva columna 'Percentage'
df_agregado['percentage'] = (df_agregado['count'] / total_students) * 100
````

````python
# Detalles

# Redondear todo a dos decimales
df_agregado = df_agregado.round(2)

# S칤mbolo de porcentaje
df_agregado['percentage'] = df_agregado['percentage'].apply(lambda x: '{:.2f}%'.format(round(x, 2)))

# Solo el primer nombre del grupo
df_agregado['race/ethnicity'] = df_agregado['race/ethnicity'].mask(df_agregado['race/ethnicity'].duplicated(), '')
````

> **Respuesta**: 
> 
Outer join: El dataframe generado por el outer join es el resultado de unir ambos dataframes completos en funci칩n de los nombres de los estudiantes. En este sentido, se logra agregar las dos columnas nuevas al dataset, pero existe un conjunto de estudiantes que estaban en un dataset y no en el otro. Es por esto que hay estudiantes que no tienen datos de g칠nero, raza, etc. En conclusi칩n, el outer join no es la mejor opci칩n para este caso.

Left join: Este dataframe es el resultado de tomar todo lo que estaba en df_grades. Esto incluyendo aquello que df_grades y other_grades compart칤an. As칤, se obtiene un dataset sin valores faltantes donde df_grades 'tom칩 aquello que le serv칤a de other grades' y se lo qued칩. 

Right join: En este caso, el resultado obtenido es de todos los datos de other_grades, junto con 'aquello que le sirve' de df_grades. As칤, dada la forma de ambos datasets, existen casos en que other_grades no puede completar las columnas de df_grades, porque este 칰ltimo no tiene datos para algunos nombres de other_grades, por lo que se obtienen espacios nulos en ciertas filas.

Inner join: El resultado de esta operaci칩n son aquellos datos que ambos datasets comparten, es decir, su intersecci칩n. As칤 se obtiene el resultado correcto sin valores nulos.

Para concluir, en este conjunto de operaciones, al realizar left join e inner join se obtiene el mismo dataset, correcto y sin casos nulos, dadas las caracter칤sticas del mismo.
