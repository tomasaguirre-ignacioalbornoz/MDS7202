````python
def generar_descripcion(fruta):
    # Documentaci√≥n Aqu√≠
    """Genera una descripci√≥n de una fruta.

    Parameters
    ----------
    fruta : dict
        Un diccionario que contiene informaci√≥n sobre la fruta.
        Debe contener las siguientes claves:
        - 'nombre' (str): El nombre de la fruta.
        - 'tipo' (str): El tipo de la fruta.
        - 'color' (str): El color de la fruta.
        - 'pepas' (bool): Un valor booleano que indica si la fruta tiene pepas o no.

    Returns
    -------
    str
        Una descripci√≥n de la fruta basada en su nombre, tipo, color y si tiene o no pepas en su interior.
    """
    # C√≥digo Aqu√≠
    if fruta.get('nombre')[-1] == 'a':
        if fruta.get('pepas') == True:
            descripcion = "La " + fruta.get('nombre') + " es una fruta de tipo " + fruta.get('tipo') + " de color " + fruta.get('color') + ". Presenta pepas en su interior." 
        elif fruta.get('pepas') == False:
            descripcion = "La " + fruta.get('nombre') + " es una fruta de tipo " + fruta.get('tipo') + " de color " + fruta.get('color') + ". No presenta pepas en su interior." 
        else: 
            descripcion = "Le falta algo al diccionario :("
    elif fruta.get('nombre')[-1] != 'a':
        if fruta.get('pepas') == True:
            descripcion = "El " + fruta.get('nombre') + " es una fruta de tipo " + fruta.get('tipo') + " de color " + fruta.get('color') + ". Presenta pepas en su interior." 
        elif fruta.get('pepas') == False:
            descripcion = "El " + fruta.get('nombre') + " es una fruta de tipo " + fruta.get('tipo') + " de color " + fruta.get('color') + ". No presenta pepas en su interior." 
        else: 
            descripcion = "Lo siento, no lo se :("
    return descripcion
            
    
````

**2. La funci√≥n `describir(datos)` que generalice la funci√≥n `generar_descripcion` para recibir una lista de frutas y devolver una lista con las descripciones de cada una:**

Por ejemplo, `describir([limon, naranja]])` deber√° devolver una lista de la forma `[descripcion_limon, descripcion_naranja]`

````python
def describir(datos):
    # Documentaci√≥n Aqu√≠
    """Generaliza la funci√≥n generar_descripcion() describiendo una lista de frutas.

    Parameters
    ----------
    datos : lista
        Una lista que contiene los diccionarios de las frutas con las claves para la funci√≥n generar_descripcion().

    Returns
    -------
    lista
        Una lista con las descripciones de las frutas.
    """
    # C√≥digo Aqu√≠
    descripciones = []
    for i in datos:
        descripciones.append(generar_descripcion(i))
    return descripciones
````

**3. La funci√≥n `filtrar_por_pepa(datos, tiene_pepas)` que dado un tipo de fruta y un booleano, retorne las frutas que tienen o no tienen pepas (seg√∫n el valor de `tiene_pepas`).**
    

Por ejemplo, `filtrar_por_pepa(datos, True)` deber√° retornar una lista con los diccionarios de: naranja, limon, frutilla y frambuesa.

````python
def filtrar_por_pepa(datos, tiene_pepas):
    # Documentaci√≥n Aqu√≠
    """Filtra las frutas introducidas seg√∫n si tienen pepas o no.

    Parameters
    ----------
    datos : lista
        Una lista que contiene los diccionarios de las frutas.
    
    tiene_pepas : boolean
        Booleano que indica si se quiere filtrar por las frutas con pepas (True) o no (False).

    Returns
    -------
    lista
        Una lista con las frutas que continen (o no) pepas.
    """
    # C√≥digo Aqu√≠
    filtrados = []
    for i in datos:
        pepas = i.get('pepas')
        if pepas == tiene_pepas:
            filtrados.append(i)
        else: pass
            
    return filtrados
````

**4. La funci√≥n `conteo_colores(datos)` que cree un diccionario que haga un conteo los colores de las frutas.**

Para estos datos, la funci√≥n deber√≠a retornar el siguiente diccionario: 
    
```python
{"amarillo": 3, "naranjo": 1, "rojo": 2}
```

````python
def conteo_colores(datos):
    # Documentaci√≥n Aqu√≠
    """Cuenta cu√°ntos colores distintos hay en las frutas ingresadas.

    Parameters
    ----------
    datos : lista
        Una lista que contiene los diccionarios de las frutas.

    Returns
    -------
    dict
        Un diccionario que indica cuantas veces se repite cada color.
    """
    # C√≥digo Aqu√≠
    conteo = {}
    for i in datos:
        if i.get('color') in conteo:
            conteo.update({i.get('color'): conteo.get(i.get('color')) + 1})
        else: 
            conteo[i.get('color')] = 1
    return conteo
````
